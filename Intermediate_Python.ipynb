```python
# @title itertools
# itertools : product . permutations,combinations,accumulate,groupby,and infinite iterators

from itertools import product # it return a cartesion product
l1 = [1,2,3]
l2 = [5,6,7,8]
prod = product(l1,l2)
print(dict(prod))  #you can print it with any iterable

print("------------------")

from itertools import permutations  #Permutations â†’ Same pairs can repeat in different orders (order matters).
# itertools.permutations(iterable, r)
per = permutations(l1,2)
print(list(per))


from itertools import combinations  #Combinations â†’ No duplicate pairs (order doesnâ€™t matter).

# itertools.combinations(iterable, r)
com = combinations(l1,2)
print(list(com))


from itertools import accumulate #give progressive accumultion
import operator
sumAccum = accumulate(l1)
print(list(sumAccum)) #every no. is the add of previos no.
mulAccum = accumulate(l1, operator.mul)
print(list(mulAccum)) #every no. is the multiply of previos no.


from itertools import groupby  #group similar elment

data = [1, 1, 2, 2, 3, 3, 3, 4, 4, 5]
grp= groupby(data)
for key , gr in grp:
  print(key,list(gr))

#  Infinite Iterators (count(), cycle(), repeat())
from itertools import count,cycle,repeat #carefully thses generate infinte sequence
print("----------")
for i in count(2,5): #start from 2 with 5 increment
  print(i)
  if i > 10:
    break
print("----------")
counter = 0
for i in cycle(['a','b','c']): #infinte over a list
  print(i)
  counter+=1
  if counter >10:
    break

print("----------")
for i in repeat("Ayush",5): #repeat the value
  print(i)

```

```python
# @title Threading vs Multiprocessing

# âœ… Threading â†’ Used for I/O-bound tasks (e.g., file reading, API calls).
from threading import Thread

def square_numbers():
    for i in range(1000):
        result = i * i


if __name__ == "__main__":
    threads = []
    num_threads = 10

    # create threads and asign a function for each thread
    for i in range(num_threads):
        thread = Thread(target=square_numbers)
        threads.append(thread)

    # start all threads
    for thread in threads:
        thread.start()

    # wait for all threads to finish
    # block the main thread until these threads are finished
    for thread in threads:
        thread.join()



# âœ… Multiprocessing â†’ Used for CPU-bound tasks (e.g., heavy calculations, ML training).
from multiprocessing import Process
import os


def square_numbers():
    for i in range(1000):
        result = i * i


if __name__ == "__main__":
    processes = []
    num_processes = os.cpu_count()

    # create processes and asign a function for each process
    for i in range(num_processes):
        process = Process(target=square_numbers)
        processes.append(process)

    # start all processes
    for process in processes:
        process.start()

    # wait for all processes to finish
    # block the main thread until these processes are finished
    for process in processes:
        process.join()

# GIL (Global Interpreter Lock) is a mechanism in Python that prevents multiple threads from running Python code at the same time.
```

```python
# @title Function Arguments

# Arguments and parameters______________
# Parameters are the variables that are defined or used inside parentheses while defining a function
# Arguments are the value passed for these parameters while calling a function
def print_name(name): # name is the parameter
    print(name)

print_name('Alex') # 'Alex' is the argument


# Positional and keyword arguments_________________
def foo(a, b, c):
    print(a, b, c)

# positional arguments
foo(1, 2, 3)

# keyword arguments
foo(a=1, b=2, c=3)
foo(c=3, b=2, a=1) # Note that the order is not important here

# mix of both
foo(1, b=2, c=3)

# default arguments________________________________________
def foo(a, b, c, d=4):
    print(a, b, c, d)

foo(1, 2, 3, 4)
foo(1, b=2, c=3, d=100)

# not allowed: default arguments must be at the end
# def foo(a, b=2, c, d=4):
#     print(a, b, c, d)

# Variable-length arguments (*args and **kwargs)_____________________________
# If you mark a parameter with one asterisk (*), you can pass any number of positional arguments to your function (Typically called *args)
# If you mark a parameter with two asterisks (**), you can pass any number of keyword arguments to this function (Typically called **kwargs).

# Forced keyword arguments___________________

# If you write '*,' in your function parameter list, all parameters after that must be passed as keyword arguments
def foo(a, b, *, c, d):
    print(a, b, c, d)

foo(1, 2, c=3, d=4)
# not allowed:
# foo(1, 2, 3, 4)


# Local vs global variables_____________________
def foo1():
    x = number # global variable can only be accessed here
    print('number in function:', x)

number = 0
foo1()

# modifying the global variable
def foo2():
    global number # global variable can now be accessed and modified
    number = 3

print('number before foo2(): ', number)
foo2() # modifies the global variable
print('number after foo2(): ', number)

# Unpacking into agruments____________________________
def foo(a, b, c):
    print(a, b, c)


# list/tuple unpacking, length must match
my_list = [4, 5, 6] # or tuple
foo(*my_list)

# dict unpacking, keys and length must match
my_dict = {'a': 1, 'b': 2, 'c': 3}
foo(**my_dict)

# my_dict = {'a': 1, 'b': 2, 'd': 3} # not possible since wrong keyword

```

```python
# @title  Shallow vs Deep Copying

# A shallow copy creates a new object, but references the same inner objects.
# If the original object contains mutable objects (like lists, dicts, etc.), modifying them in one copy affects the other.

import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow = copy.copy(original)  # Creates a shallow copy

shallow[0][0] = 99  # Modifying the first element

print("Original:", original)  # Also modified!
print("Shallow:", shallow)


# A deep copy creates a completely independent copy, including all nested objects.
# Modifying one does not affect the other.
import copy

original = [[1, 2, 3], [4, 5, 6]]
deep = copy.deepcopy(original)  # Creates a deep copy

deep[0][0] = 99  # Modifying the first element

print("Original:", original)  # Not modified!
print("Deep:", deep)


```
